```python
df = pd.read_csv('../input/kiva_loans.csv')
df[:3]
```

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
le = preprocessing.LabelEncoder().fit(df['date'])
df['date'] = le.transform(df['date'])
```

```python
x = df[['date', 'loan_amount']]
y = df['funded_amount']

xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 0.4)
```

```python
from sklearn.linear_model import LinearRegression
linReg = LinearRegression()
linReg.fit(xTrain, yTrain)
```

```python
predictions = linReg.predict(xTest)
from sklearn import metrics
np.sqrt(metrics.mean_squared_error(predictions, yTest))
```
